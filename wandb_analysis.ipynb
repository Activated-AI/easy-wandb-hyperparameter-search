# Analyzing Individual Runs

```python
# url for run we want to analyze:
# https://wandb.ai/llm-research-activated/pytorch-sweeps-demo/runs/s63f4228
entity = 'llm-research-activated'
project = 'pytorch-sweeps-demo'
run_id = 's63f4228'

# url for sweep we want to analyze:
# https://wandb.ai/llm-research-activated/pytorch-sweeps-demo/sweeps/blm9j8cg?nw=nwuserjcaples4
sweep_id = 'blm9j8cg'
```

```python
import wandb
from pprint import pprint
api = wandb.Api()
wandb.login()

run = api.run(f"{entity}/{project}/{run_id}")
```

```python
pprint(run.summary_metrics)
```

```python
history = run.history()
losses = history["batch loss"]
```

```python
import matplotlib.pyplot as plt

# Assuming losses is already defined
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Losses time')
plt.show()
```

# Analyzing the sweep

```python
api = wandb.Api()
sweep = api.sweep(f"{entity}/{project}/{sweep_id}")
runs = sweep.runs
```

```python
best_run = sweep.best_run()
```
